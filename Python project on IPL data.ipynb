```python
#loading the required libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#loading the data set and have a view of first 5 records
IPL=pd.read_csv('matches.csv')
IPL.head()
```

```python
#Checking the no. of rows and columns
IPL.shape
```

```python
# count of the man of the match for each player
IPL['player_of_match'].value_counts()
```

```python
#count of top 10 man of the match
IPL['player_of_match'].value_counts()[:10]
```

```python
# getting the top 5 players name(only) with most man of the match awards
list(IPL['player_of_match'].value_counts()[:5].keys())
```

```python
#plotting a barplot for the top 5 players with most man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(IPL['player_of_match'].value_counts()[:5].keys()),list(IPL['player_of_match'].value_counts()[:5]),
            color='skyblue')
plt.show()
```

```python
# getting the count of result
IPL['result'].value_counts()
```

```python
#finding the count of toss winner
IPL['toss_winner'].value_counts()
```

```python
# extracting the records where a team won on batting first
batting_first=IPL[IPL['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#Histogram for win by run
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
# bar plot for top 3 teams with most wins on batting first
plt.figure(figsize=(6,4))
plt.bar(list(batting_first['winner'].value_counts()[:3].keys()),list(batting_first['winner'].value_counts()[:3]),
       color=['turquoise','skyblue','yellow'])
plt.show()
```

```python
# piechart for winner by batting first
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# extracting those records where a team has won by batting second
batting_second=IPL[IPL['win_by_wickets']!=0]
batting_second.head()
```

```python
# histogram for teams win by wickets on batting second
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=20)
plt.show()
```

```python
# most wins on batting second
batting_second['winner'].value_counts()
```

```python
# barplot for most wins on batting second
plt.figure(figsize=(6,4))
plt.bar(list(batting_second['winner'].value_counts()[:3].keys()),list(batting_second['winner'].value_counts()[:3]),
        color=['r','g','b'])
plt.show()
```

```python
# piechart with most wins on batting second
plt.figure(figsize=(6,6))
plt.pie(list(batting_second['winner'].value_counts()),autopct='%0.1f%%',labels=list(batting_second['winner'].value_counts().keys()))
plt.show()
```
