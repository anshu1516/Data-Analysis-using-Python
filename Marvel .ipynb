```python
import pandas as pd
from matplotlib import pyplot as plt
marvel=pd.read_csv('charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
good=marvel[marvel['Alignment']=='good']
good.head()
```

```python
good.sort_values(by='Speed',ascending=False).head()
```

```python
good.sort_values(by='Power',ascending=False).head()
```

```python
max_power_full=good[good['Power']==100]
max_power_full
```

```python
max_power_full.shape
```

```python
max_power=good.sort_values(by='Total',ascending=False)
max_power.head()
```

```python
plt.figure(figsize=(8,6))
plt.bar((max_power['Name'][:5]),(max_power['Total'][:5]),color='g')
plt.show()
```
