```python
import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
df.head()
```

```python
df.tail()
```

```python
# keeping only required columns
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
# rename the columns
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
# looking for lastest dates
df.tail()
```

```python
# current date situation
today=df[df.date=='2021-05-17']
today.head()
```

```python
# sorting data w.r.t number of confirmed cases
max_confirmed=today.sort_values(by='confirmed',ascending=False)
max_confirmed
```

```python
# getting states with maximum number of confirmed cases
top_confirmed=max_confirmed[:5]
top_confirmed
```

```python
# making barplot for states with top confirmed cases
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x='state',y='confirmed',hue='state',data=top_confirmed)
plt.show()
```

```python
# sorting data w.r.t number of deaths
max_death=today.sort_values(by='deaths',ascending=False)
max_death
```

```python
top_death=max_death[:5]
top_death
```

```python
# making barplot for top 5 states with most death cases 
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x='state',y='deaths',hue='state',data=top_death)
plt.show()
```

```python
# sorting data w.r.t number of cured cases
max_cured=today.sort_values(by='cured',ascending=False)
max_cured
```

```python
top_cured=max_cured[:5]
top_cured
```

```python
# making barplot for top 5 states with most number of cured cases
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x='state',y='cured',hue='state',data=top_cured)
plt.show()
```
